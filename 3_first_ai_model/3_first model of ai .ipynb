<h1 style = "font-size:3rem;color:purple;">Normalize training data before training the neural net</h1>

```python
x_train = x_train/255 # it is the another way to normalize data like: Minmaxscaler and StanderdScaler
x_test = x_test/255 
```

<h1 style = "font-size:3rem;color:purple;">Now build the Sequential Model and add layers into it </h1>

```python
from keras.models import Sequential 
from keras.layers import Flatten, Dense, Activation 
```

```python
model = Sequential()
model.add(Flatten(input_shape = [28,28])) # it flates the 2d to 1d array 
model.add(Dense(20,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

    This code defines a simple neural network using Keras with the Sequential API. Here's a step-by-step explanation of each line:

1. model = Sequential():

-        This initializes a sequential model, which means the layers will be added to the model one after the other, in a sequence.The Sequential model is a linear stack of layers. This is suitable for a plain feed-forward neural network where data flows from one layer to the next without any complex branching or looping.


2. model.add(Flatten(input_shape=[28,28])):

-        The Flatten layer reshapes the 2D input array into a 1D vector.In this case, the input is a 2D image of shape 28x28 pixels (typically a grayscale image like MNIST digits).Flattening converts this 28x28 array into a 1D array with 784 elements (28 * 28 = 784).The input_shape=[28, 28] specifies the shape of the input data. This is only needed for the first layer in the model.


3. model.add(Dense(20, activation='relu')):

-        This adds a Dense (fully connected) layer with 20 neurons.Each neuron in this layer is connected to all neurons in the previous layer.The ReLU (Rectified Linear Unit) activation function is applied to the output of this layer. ReLU is a common activation function used in neural networks, which outputs the input directly if it's positive and zero otherwise (f(x) = max(0, x)).The output of this layer will be a 1D array with 20 elements.


4. model.add(Dense(10, activation='softmax')):

-        This adds another Dense layer with 10 neurons. This layer is typically used for classification tasks where there are 10 classes (e.g., in the MNIST dataset, where there are 10 digits from 0 to 9).The softmax activation function is used, which converts the raw output values into probabilities that sum to 1. Each output will represent the probability of the input belonging to one of the 10 classes.

- Or (Similar code upper one and lower one)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense

model = Sequential([
    Flatten(input_shape=(28, 28)),  # Direct input_shape in layer
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
model.fit(x_train,y_train)
```

```python
model.evaluate(x_test,y_test)
```

## Above shows accuracy score of 85.69%. The first parameter is loss
