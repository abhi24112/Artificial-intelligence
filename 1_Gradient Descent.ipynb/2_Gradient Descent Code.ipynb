```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):

    # we have to select any initial value to start the process
    m_curr = b_curr = 0 
    iteration = 1000
    n = len(x)
    learning_rate = 0.06 # Learning Rate()alpha (we have to do hit and trial to find LR)
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='blue', marker='*', s=100, label='Data Points')
    for i in range(iteration):
        # predicted value of (y)
        y_pre = m_curr * x + b_curr
        # Plot the current line
        #cost function (mse)
        cost = (1/n) * sum((y-y_pre)**2)

        # partial differentiation of (mse) with resp. of (m) and (b)
        md = -(2/n) * sum(x*(y-y_pre))
        bd = -(2/n) * sum((y-y_pre))

        # Learning Rate 
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        # print(f"m:{m_curr} b:{b_curr} cost:{cost} iteration:{i+1}")
        # if i == 0:
        #     plt.plot(x, y_pre, color='green', label='Fit Line')
        if i == iteration-1:
            plt.plot(x, y_pre, color='red',linewidth=2, label='Best Fit Line') # optimized m and b
            print(f"m:{m_curr} b:{b_curr} cost:{cost} iteration:{i+1}")
            
        else:
            plt.plot(x,y_pre, color='green')
        
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gradient Descent Visualization')
    plt.grid()
    plt.legend()
    plt.show()
    
# Example data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
gradient_descent(x, y)
```
